# Improving Plot Aesthetics

The dataset contains the percentage of bachelor's degrees granted to women from 1970 to 2012. It is broken up into 17 categories of degrees, with each column as a separate category.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
women_degrees.head()
```

```python
plt.plot(women_degrees['Year'], women_degrees['Biology'])
```

```python
plt.plot(women_degrees['Year'],women_degrees['Biology'],color='b',label='women')
plt.plot(women_degrees['Year'],100-women_degrees['Biology'],color='r',label='men')
plt.legend(loc="upper right")
plt.title("Percentage of Biology Degrees Awarded By Gender")
```

```python
fig, ax = plt.subplots()
ax.plot(women_degrees['Year'],women_degrees['Biology'],color='b',label='women')
ax.plot(women_degrees['Year'],100-women_degrees['Biology'],color='r',label='men')
ax.tick_params(bottom="off", top="off", left="off", right="off")
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc="upper right")
```

```python
print(ax.spines)
```

```python
fig, ax = plt.subplots()
ax.plot(women_degrees['Year'],women_degrees['Biology'],color='b',label='women')
ax.plot(women_degrees['Year'],100-women_degrees['Biology'],color='r',label='men')
ax.tick_params(bottom="off", top="off", left="off", right="off")
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc="upper right")
for key, spine in ax.spines.items():
    spine.set_visible(False)
```

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = plt.figure(figsize=(12, 12))

for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c='blue', label='Women')
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c='green', label='Men')
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(major_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")

    plt.legend(loc='upper right')
plt.show()
```
