# Multiple Plots

The dataset contains the monthly unemployment rate as a CSV from January 1948 to August 2016.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
unrate = pd.read_csv("unrate.csv")
unrate.head()
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
plt.plot(unrate['DATE'][:12],unrate['VALUE'][:12])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
plt.show()
```

```python
# Grid positioning
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate['DATE'][0:12],unrate['VALUE'][0:12])
ax2.plot(unrate['DATE'][12:24],unrate['VALUE'][12:24])
plt.show()
```

```python
# Grid positioning
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(unrate['DATE'][0:12],unrate['VALUE'][0:12])
ax1.set_title("Monthly Unemployment Rate, 1948")
ax2.plot(unrate['DATE'][12:24],unrate['VALUE'][12:24])
ax2.set_title("Monthly Unemployment Rate, 1949")
plt.show()
```

```python
fig = plt.figure(figsize=(12,18))

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    starting_index = i*12
    ending_index = (i+1)*12
    subset = unrate[starting_index:ending_index]
    ax.plot(subset['DATE'],subset['VALUE'])
    ax.set_title("Monthly Unemployment Rate"+str(1948+i))
plt.show()
```

```python
# Overlaying Line Charts
unrate["Month"] = unrate['DATE'].dt.month
```

```python
unrate.head()
```

```python
fig = plt.figure(figsize=(6,3))
plt.plot(unrate["Month"][0:12],unrate["VALUE"][0:12],c='blue')
plt.plot(unrate["Month"][12:24],unrate["VALUE"][12:24],c='red')
plt.show()
```

```python
colors = ["red","blue","orange","green","black"]
fig = plt.figure(figsize=(10,6))
for i in range(5):
    starting_index = i*12
    ending_index = (i+1)*12
    subset = unrate[starting_index:ending_index]
    plt.plot(subset["Month"],subset["VALUE"],label = str(1948+i),c=colors[i])
plt.legend(loc = "upper left")
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
